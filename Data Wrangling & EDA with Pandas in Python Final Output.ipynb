# TASK #5: PERFORM NORMALIZATION

```python
# Feature Scaling is an important step to take prior to training of machine learning models to ensure that features
# are within the same scale. Example: Interest rate and employment score are at a different scale. This will result
# in one feature dominating the other feature. Scikit Learn offers several tools to perform feature scaling.

# Like MinMaxScaler(), StandardScaler(), MaxAbsScaler(), RobustScaler(), and Normalizer(). All have their own use.
# Refer to the documentations for that. But this time we will use the MinMaxScaler() which is best for most cases.

# Normalization is conducted to make feature values range from 0 to 1.
```

```python
# Let's read a CSV file using Pandas as follows
df = pd.read_csv('Life_Expectancy_Data.csv')
df
```

```python
df['Life expectancy '].values
```

```python
# Normalization is conducted to make feature values range from 0 to 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Life expectancy '] = scaler.fit_transform(df['Life expectancy '].values.reshape(-1,1))
```

# TASK #6: PERFORM STANDARDIZATION

```python
# Another Feature Scaling strategy is called Standardization. It is conducted to transfrom the data to have a mean
# of 0, and standard deviation of 1. Standardization is also known as Z-score normalization in which properties
# will have the behaviour of a standard normal distribution.

# Z = (x - xbar)/sigma

# ALWAYS REMEMBER!
# A normalized dataset will always range from 0 to 1 while a Standardized dataset will always have a mean of 0 and
# standard deviation of 1, but can have any upper and lower values.
```

```python
# Let's read a CSV file using Pandas as follows
df = pd.read_csv('Life_Expectancy_Data.csv')
df
```
